```python
def decorator(func):
    def wrapper(*args):
        print('hi')
        func(*args)
    return wrapper
```

```python
@decorator
def sayhi():
    print('helloooo')
```

```python
sayhi()
```

```python
@decorator
def greet(a):
    print(a*10)
```

```python
greet('a')
```

```python
def deco(func):
    def inner(*args):
        b = func(*args).upper()
        print(b)
    return inner
```

```python
@deco
def greet(a):
    return a
```

```python
greet('hi')
```

```python
def dec(func):
    def inner(*args):
        a = func(*args)
        print(a*10)
    return inner

```

```python
@deco
def test(*args):
    for a in args:
        print('hi')
    return 'hello'
```
